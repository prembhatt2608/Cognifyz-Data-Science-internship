```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\cognifyz internship\Dataset .csv")
```

```python
plt.hist(df['Aggregate rating'], bins=20)
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
df['Aggregate rating'].value_counts().sort_index().plot(kind='bar')
plt.title("Rating Frequency")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
df_cuisine = df.copy()
df_cuisine['Cuisines'] = df_cuisine['Cuisines'].str.split(', ')
df_cuisine = df_cuisine.explode('Cuisines')
```

```python
cuisine_rating = df_cuisine.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)

cuisine_rating.head(10).plot(kind='bar')
plt.title("Top 10 Cuisines by Average Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
city_rating = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)

city_rating.head(10).plot(kind='bar')
plt.title("Top Cities by Average Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(df['Votes'], df['Aggregate rating'])
plt.xlabel("Votes")
plt.ylabel("Rating")
plt.title("Votes vs Rating")
plt.show()
```

```python
df.groupby('Price range')['Aggregate rating'].mean().plot(kind='bar')
plt.title("Price Range vs Average Rating")
plt.xlabel("Price Range")
plt.ylabel("Rating")
plt.show()
```

```python
plt.scatter(df['Average Cost for two'], df['Aggregate rating'])
plt.xlabel("Average Cost for Two")
plt.ylabel("Rating")
plt.title("Cost vs Rating")
plt.show()
```

```python
import seaborn as sns

corr = df[['Aggregate rating', 'Votes', 'Average Cost for two', 'Price range']].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x='Price range', y='Aggregate rating', data=df)
plt.title("Price Range vs Rating Distribution")
plt.show()
```
